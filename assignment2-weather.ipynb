## Weather reading.


Import python packages.

```python

import matplotlib.pyplot as plt # For plotting.

from matplotlib.pyplot import figure

import pandas as pd # For data visualisation.

import numpy as np # For numerical operations on arrays.
```

Loading Data


```python
dataFrame = pd.read_csv(r"C:\Users\phumi\OneDrive\Desktop\PFDA-courseware\assignment\weatherreadings1.csv")
```

Reading in  Data

```python
dataFrame.head(3)
```

Selecting Date and Temperature columns.

```python
# variables for plotting the graph
result = dataFrame[['createdDateTime', 'dryBulbTemperature_Celsius']]
result.head(3)
```

Plotting

```python
x = dataFrame['createdDateTime']
y = dataFrame['dryBulbTemperature_Celsius']

plt.xticks(np.arange(0, len(x)+1, 3)) # change x-axis tick spacing.
plt.xticks(rotation=90) # rotate  x-axis labels.

plt.tick_params(axis='x', which='major', labelsize=7) # adjust x-axis label size.

plt.plot(x, y)
plt.show()
```

Adding Labels to the plot.

```python


plt.xlabel('Date & Time') # adding labels to x-axis
plt.ylabel('Temperature [C]') # adding labels to y-axis

plt.xticks(np.arange(0, len(x)+1,3)) # change x-axis tick spacing.
plt.xticks(rotation=90) # rotate x-axis label.

plt.tick_params(axis='x', which='major', labelsize=7) # adjust x-axis label size
plt.title("Temperature vs Date Plot") # Title


plt.plot(x, y, 'r.-')

```
